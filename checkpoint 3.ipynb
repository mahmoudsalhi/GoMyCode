```python
#Question 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ds=pd.read_csv("titanic-passengers.csv", sep=';')
ds.head()
ds.isnull().sum()
```

```python
ds['Cabin'].fillna('G6',inplace=True)
ds['Age'].fillna(ds['Age'].mean(), inplace=True)
ds['Embarked'].fillna('S',inplace=True)
```

```python
def plot_correlation_map(ds):
    corr = ds.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 } 
        )
plot_correlation_map(ds)
```

```python
# this is a heatmap : it contains values representing various shades of the same colour for each value to be plotted
```
